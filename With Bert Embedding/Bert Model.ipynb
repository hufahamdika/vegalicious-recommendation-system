```python
import json
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
with open('full_format_recipes.json', 'r') as f:
  data = json.load(f)
data[0]
```

```python
categories = []
for data_temp in data:
  if 'categories' in data_temp:
    for category in data_temp['categories']:
      if category not in categories:
        categories.append(category)
```

```python
print(len(data))
```

```python
include_categories = ['Vegan']
data_vegan = []
for data_temp in data:
  if 'categories' in data_temp:
    for category in data_temp['categories']:
      if category in include_categories:
        data_vegan.append(data_temp)
```

```python
print(len(data_vegan))
```

```python
categories_vegan = {}
for data_temp in data_vegan:
  if 'categories' in data_temp:
    for category in data_temp['categories']:
      if category not in categories_vegan:
        categories_vegan[category] = 1
      else:
        categories_vegan[category] += 1
```

```python
categories_vegan = sorted(categories_vegan.items(), key=lambda x: x[1], reverse=True)
```

```python
categories_vegan[:20]
```

```python
print(len(categories_vegan))
```

```python
df_all = pd.DataFrame(data, columns=data[0].keys())
df_all
```

```python
df_vegan = pd.DataFrame(data_vegan, columns=data_vegan[0].keys())
df_vegan
```

```python
df_all = df_all.drop('date', axis=1)
df_vegan = df_vegan.drop('date', axis=1)

df_all = df_all.dropna(subset=['title', 'categories'])
df_vegan = df_vegan.dropna(subset=['title', 'categories'])
```

```python
df_all['is_vegan'] = df_all['categories'].apply(lambda x: 1 if 'Vegan' in x else 0)
df_all
```

```python
convert_dict = {'directions': str,
                'fat': np.float32,
                'categories': str,
                'ingredients': str,
                'calories': np.float32,
                'protein': np.float32,
                'rating': np.float32,
                'sodium': np.float32,
                }
```

```python
df_all = df_all.astype(convert_dict)
df_vegan = df_vegan.astype(convert_dict)

df_all = df_all.drop_duplicates()
df_vegan = df_vegan.drop_duplicates()
```

```python
print(f'length all data = {len(df_all)}\n')
print(f'null all data = \n{df_all.isnull().sum()}\n')
print(f'length vegan data = {len(df_vegan)}\n')
print(f'null vegan data = \n{df_vegan.isnull().sum()}\n')
```

```python
import pandas as pd
import re
import tensorflow_hub as hub
import tensorflow as tf
import tensorflow_text as text
from sklearn import metrics
```

# Training All Data

```python
preprocessor = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/small_bert/bert_en_uncased_L-2_H-128_A-2/1", trainable=True)

text_input = tf.keras.layers.Input(shape=(), dtype=tf.string)
x = preprocessor(text_input)
x = encoder(x)
embedding_model_all = tf.keras.Model(text_input, x['pooled_output'])
embedding_model_all.compile()
```

```python
embedding_model_all.summary()
```

```python
df_all['encoding'] = df_all['title'].apply(lambda x: embedding_model_all(tf.constant([x])))
```

```python
df_all
```

```python
df_all['encoding'].iloc[1]
df_all['encoding'].shape
```

```python
df_all_trained = df_all.copy()
df_all_trained
```

```python
df_all_trained['is_vegan'].value_counts()
```

```python
df_all_trained_vegancut = df_all_trained.copy()
df_all_trained_vegancut = df_all_trained_vegancut[df_all_trained_vegancut['is_vegan'] == 1]
df_all_trained_vegancut = df_all_trained_vegancut.dropna()
df_all_trained_vegancut = df_all_trained_vegancut.reset_index()
df_all_trained_vegancut = df_all_trained_vegancut.drop(columns=['index'])
```

```python
def preprocess_text(text):
  text = text.lower()
  text = re.sub('[^A-Za-z0-9]+', ' ', text)
  return tf.constant([text])
```

```python
test = df_vegan[['title', 'protein', 'desc']].iloc[50]
sentences = test['title']
sentences
```

```python
query_text = preprocess_text(sentences)
query_encoding = embedding_model_all(query_text)

df_all_trained_vegancut['similarity_score'] = df_all_trained_vegancut['encoding'].apply(lambda x: metrics.pairwise.cosine_similarity(x, query_encoding)[0][0])
df_all_trained_vegancut = df_all_trained_vegancut.sort_values(by=['similarity_score'], ascending=False)
df_all_trained_vegancut[:20]
```

```python
embedding_model_all.save('get_bert_embeddings_alltrained.h5')
```
